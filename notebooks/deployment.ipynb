```python
import dash
import dash_bootstrap_components as dbc
import dash_html_components as html
import dash_core_components as dcc

app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

app.layout = dbc.Container(
    style={
        'background-image': 'url("/assets/image.jpg")',
        'background-size': 'cover',
        'background-position': 'center',
        'position': 'fixed',
        'top': 0,
        'left': 0,
        'width': '100%',
        'height': '100%',
        'z-index': '-1',
        'textAlign': 'center',
        'marginTop': '0px',
    },
    fluid=True,
    children=[
        dbc.Row(
            dbc.Col(
                html.H1(
                    'Mobile Price Classification',
                    style={
                        'background-color': 'lightgray',
                        'color': 'black',
                        'display': 'inline-block',
                        'padding': '1px'
                    }
                )
            )
        ),
        dbc.Row(
            dbc.Col(
                html.H1(
                    'ATW Ltd',
                    style={
                        'textAlign': 'center',
                        'background-color': 'lightgray',
                        'color': 'black',
                        'display': 'inline-block',
                        'padding': '1px',
                        'fontSize': '20px'
                    }
                )
            )
        ),
        dbc.Row(
            [
                dbc.Row(
                            dbc.Col(
                                html.Div(style={'height': '80px'})  # Add an empty div with a fixed height
                            )
                        ),
                dbc.Col(
                    [
                        dbc.Row(
                            dbc.Col(
                                html.Div(style={'height': '100px'})  # Add an empty div with a fixed height
                            )
                        ),
                        dbc.Row(
                            dbc.Col(
                                html.Div(
                                    dbc.Input(
                                        type='list',
                                        placeholder = 'Enter the test example',
                                        id='input',
                                        style={'width': '300px','height':'35px', 'marginLeft': '0px', 'textAlign': 'center','backgroundColor': 'lightgray'}
                                    ),
                                    style={'marginLeft': '18px'}
                                )
                            )
                        ),
                        dbc.Row(
                            dbc.Col(
                                html.Div(style={'height': '100px'})  # Add an empty div with a fixed height
                            )
                        ),
                        dbc.Row(
                            dbc.Col(
                                html.Div(
                                    id='output-container',
                                    children=[
                                        dbc.Alert(
                                            id='output',
                                            color='success',
                                            style={'textAlign': 'center'}
                                        )
                                    ],
                                    style={'marginTop': '20px', 'fontSize': '30px','background-color': 'lightgray', 'padding': '0px','display': 'inline-block'}
                                )
                            )
                        )
                    ]
                )
            ]
        )
    ]
)





@app.callback(
    dash.dependencies.Output('output', 'children'),
    [dash.dependencies.Input('input', 'value')]
)


def update_output(value):
    # Check if the value is not None
    if value is not None:
        try:
            # Convert the string value to a list of floats
            value_list = ast.literal_eval(value)

            # Make the prediction using the model
            predictions = model.predict([value_list])

            # Create the text with the predicted category
            txt = f"The mobile Price Range is in category: {predictions[0]}"

            # Return the output to be displayed
            return txt
        except (ValueError, SyntaxError):
            # Handle any error that may occur during evaluation
            return 'Invalid input. Please enter a valid list of values.'
    else:
        return 'No input provided.'


if __name__ == '__main__':
    app.run_server(debug=False)
```
